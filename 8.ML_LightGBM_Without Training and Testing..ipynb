```python
import pandas as pd
import warnings
warnings.filterwarnings("ignore")
```

```python
dataset = pd.read_csv("insurance_pre.csv")
dataset
```

```python
dataset = pd.get_dummies(dataset,drop_first = True)
dataset
```

```python
dataset.columns
```

```python
independent = dataset[['age', 'bmi', 'children', 'sex_male', 'smoker_yes']]
independent
```

```python
dependent = dataset[['charges']]
dependent
```

```python
from sklearn.model_selection import GridSearchCV
from lightgbm import LGBMRegressor # procedure
param_grid = {'n_estimators' :[100,200,500,1000],'learning_rate':[0.01,0.03,0.05,0.1],'max_depth':[5,10,15,20]}
grid = GridSearchCV(LGBMRegressor(),param_grid, refit=True, verbose=3, n_jobs=1)
grid.fit(independent,dependent) 
```

```python
re = grid.cv_results_
print("The R Score Value for Best Parameter {}:".format(grid.best_params_))
```

```python
table = pd.DataFrame.from_dict(re)
table
```

```python
age_input = float(input("Age :"))
bmi_input = float(input("BMI :"))
children_input = float(input("Children :"))
sex_male_input = int(input("Sex Male 0 or 1 :"))
smoker_yes_input = int(input("Smoker_yes 0 or 1 :"))
```

```python
Future_Prediction = grid.predict([[age_input,bmi_input,children_input,sex_male_input,smoker_yes_input]])
print("Future_Prediction {}:".format(Future_Prediction))
```
