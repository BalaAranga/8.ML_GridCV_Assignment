```python
import pandas as pd
import warnings
warnings.filterwarnings("ignore")
```

```python
dataset = pd.read_csv("insurance_pre.csv")
dataset
```

```python
dataset = pd.get_dummies(dataset,drop_first = True)
dataset
```

```python
dataset.columns
```

```python
independent = dataset[['age', 'bmi', 'children', 'sex_male', 'smoker_yes']]
independent
```

```python
dependent = dataset[['charges']]
dependent
```

```python
from sklearn.model_selection import GridSearchCV
from xgboost import XGBRegressor # procedure
param_grid = {'eta' :[0.1,0.3,1,1.5],'gamma':[50,100,200,500]}
grid = GridSearchCV(XGBRegressor(),param_grid, refit=True, verbose=3, n_jobs=1)
grid.fit(independent,dependent) 
```

```python
re = grid.cv_results_
print("The R Score Value for Best Parameter {}:".format(grid.best_params_))
```

```python
table = pd.DataFrame.from_dict(re)
table
```
