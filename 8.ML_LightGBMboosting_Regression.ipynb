```python
import pandas as pd
```

```python
dataset = pd.read_csv("insurance_pre.csv")
dataset
```

```python
dataset = pd.get_dummies(dataset,drop_first = True)
dataset
```

```python
dataset.columns
```

```python
independent = dataset[['age', 'bmi', 'children', 'sex_male', 'smoker_yes']]
independent
```

```python
dependent = dataset[['charges']]
dependent
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(independent,dependent,test_size=0.30,random_state=0) 
```

```python
# Model Creation
# important parameters
# n_estimators use between 100 and 1000
# learning_rate use between 0.01 and 0.1
# max_depth use between 5 and 20
from lightgbm import LGBMRegressor # procedure
regressor = LGBMRegressor(n_estimators= 100,learning_rate = 0.05,max_depth = 5)
regressor = regressor.fit(x_train,y_train)  # Model created
```

```python
y_pred = regressor.predict(x_test) # AI Brain giving output(predicting)
y_pred
```

```python
from sklearn.metrics import r2_score
r_score = r2_score(y_test,y_pred)
```
